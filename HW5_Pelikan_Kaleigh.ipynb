```r
Data_Frame <- data.frame(
    id = c(1,2,3,4,5),
    name = c("Peter", "Amy", "Ryan", "Gary", "Michelle"),
    salary = c(623.30, 515.20, 611.00, 729.00, 843.25)
)

Data_Frame
```

```r
new_col_DF <- cbind(Data_Frame, "department" = c("statistics", "computer science", "marketing", 
                                                 "finance", "business analytics"))
new_col_DF
```

```r
Data_Frame_New <- new_col_DF[-c(1,3,5), -c(2,3)]
Data_Frame_New
```

```r
x <- c("Peter", "Gary", "Michelle")
y <- c(623.30, 729.00, 843.25)

barplot(y, names.arg = x, density = 10, main = "Salaries for Different Departments", 
        xlab = "salary", ylab = "name")
```

```r
mylabel <- c("Maximum Salary", "Minimum Salary", "Median Salary")

colors <- c("lightblue", "lightpink", "orange")

x <- c(843.25, 515.20, 623.3)

pie(x, label = mylabel, main = "Salary Statistics", col = colors)

legend("bottomright", mylabel, fill = colors)
```
